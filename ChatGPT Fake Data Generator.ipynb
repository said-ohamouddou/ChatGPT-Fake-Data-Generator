```python
import openai
import pandas as pd
import json
import ast

class ChatGptFakeDataGenerator:
    def __init__(self, features_list,rows_number,openai_key):
        self.features_list = features_list
        self.rows_number=rows_number
        self.openai_key=openai_key
        self.df=None
        self.generate_data()

    def get_dataframe(self):
        return self.df
            
    def get_json(self):
        return self.df.to_json()
    
    def generate_data(self):
        data_list = []
        try:
            openai.api_key = self.openai_key
            basic_query= "fake data for"+ str(self.rows_number)+ "rows with the features of "+str(self.features_list)+" in a key-value data structure format"
            answer = openai.ChatCompletion.create(model='gpt-3.5-turbo', messages=[{'role': 'user', 'content': basic_query}])
            answer = answer['choices'][0]['message']['content'].replace('\n', '')
            answer_dict=ast.literal_eval(answer)
            self.df= pd.DataFrame.from_dict(answer_dict)
            
        except:
               print("We cannot generate the data check your inputs!")
```
